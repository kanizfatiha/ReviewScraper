```python
# Step 1: Open the file in reading mode
file_path = '/content/Links-Square.txt'
try:
    with open(file_path, 'r') as file:

        # Step 2: Read the content
        links = file.readlines()  # Reads all lines into a list

        # Process and print the content
        print(links)

# Handle file not found or other exceptions
except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print("An error occurred:", e)
```

```python
import re

item_ids = []

# Loop through each URL and extract the item ID
for url in links:
    match = re.search(r'-i(\d+)-', url)
    if match:
        item_id = str(match.group(1))
        item_ids.append(item_id)

# Print the extracted item IDs
print("Extracted Item IDs:", item_ids)
```

```python
import requests
import re
import json

headers = {
    'authority': 'www.daraz.com.bd',
    'pragma': 'no-cache',
    'cache-control': 'no-cache',
    # ... (other headers)
}

# Set your item ID, page size, and initial page number
page_size = 10
page_no = 1

review_data = []

scraped_reviews = []  # To store the scraped reviews
for item_id in item_ids:
    page_no = 1  # Start from the first page for each item
    while True:
        review_url = f'https://my.daraz.com.bd/pdp/review/getReviewList?itemId={item_id}&pageSize={page_size}&filter=0&sort=0&pageNo={page_no}'
        review_request = requests.get(review_url, headers=headers)

        if review_request.status_code != 200:
            print(f"Reached the end of pagination for item {item_id} at page {page_no - 1}")
            break

        try:
            review_data = review_request.json()
            reviews_on_page = review_data.get('model', {}).get('items', [])
            if not reviews_on_page:
                print(f"No more reviews for item {item_id} on page {page_no}")
                break

            # Extract and process the reviews here
            scraped_reviews.extend(reviews_on_page)

            print(f"Scraped reviews for item {item_id} from page {page_no}")
            page_no += 1

        except json.JSONDecodeError as e:
            print(f"Error decoding JSON for item {item_id} on page {page_no}: error {e}")
            break


# Now you have scraped reviews from all pages
# Process or save the reviews as needed
```
